By checking some the columns using the boxplot we can see the presence of outliers in variuos columns with thier range and shape in the columns, now we can proceed to treat the outlies using IQR method, replacing the upperside outliers with upper quartile range and lowerside outliers with lower quartile using function

### Outliers detection and fixes

```python
def outliers_detection_fixes(df_cleaned,columns):
    for col in columns:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        Lower = Q1 - 1.5 * IQR
        Upper = Q3 + 1.5 * IQR
        df_cleaned[col] = np.where(df_cleaned[col] > Upper, Upper, np.where(df_cleaned[col] < Lower, Lower, df_cleaned[col]))
    return df_cleaned
```

```python
outliers_detection_fixes(df_cleaned, df_cleaned.columns)
```

### Normalization / Standardization

```python
df_cleaned.describe().T
```

From the max parameter of each column, we can see there is different range of the data present in each columns, so to get the data in scaled form we use the techniques normalization or standardization. 
The MinMaxScaler transforms data in between range 0 to 1 and Standard Scaler transforms data in between thr range -1 to +1 assuming the data is normally distributed.
Both the techniques are sentive to outliers, so before normalization the outliers needs to be treated to get the proper scaled matrix of the data with normalisation.

```python
## Normalization of the data
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler

mm_scaler= MinMaxScaler(feature_range=(0,1))
rescaled_mm = mm_scaler.fit_transform(df_cleaned)
```

```python
## Normalised data using MinMaxscaler
set_printoptions(precision=2)
print(rescaled_mm[0:5,:])
```

```python
## Standardizing Data (mean = 0 , std = 1)
from sklearn.preprocessing import StandardScaler

sd_scaler = StandardScaler()
rescaled_sd = sd_scaler.fit_transform(df_cleaned)
```

```python
## Standardized data using StandardScaler
set_printoptions(precision=2)
print(rescaled_sd[0:5,:])
```

```python
## Converting scaled data into dataframe 
pd.DataFrame(rescaled_sd)
```

```python
## Assigning column labels to scaled data 
scaled_df = pd.DataFrame(rescaled_sd, columns=df_cleaned.columns)
scaled_df
```

The scaled data is ready to use for model building, we can use this scaled data to build the different ML models using the proper algorithms
